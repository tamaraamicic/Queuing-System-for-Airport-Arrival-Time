```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BASA = pd.read_csv("BASA_AUC_2028_912.csv")
```

```python
BASA
```

```python
daily_avg_wait_time_by_day_of_week = BASA.groupby('Day_of_Week')['Wait_Time'].mean()
daily_avg_wait_time_by_day_of_week.plot(kind = 'line')
plt.xlabel('Day of Week')
plt.ylabel('Average Wait Time')
plt.title('Average Wait Time by Day of Week')
plt.show()
```

```python
daily_avg_wait_time_by_month = BASA.groupby('Month')['Wait_Time'].mean()
daily_avg_wait_time_by_month.plot(kind = 'line')
plt.xlabel('Month')
plt.ylabel('Average Wait Time')
plt.title('Average Wait Time by Month')
plt.show()
```

```python
daily_avg_wait_time_by_time_of_day = BASA.groupby('Time_of_Day')['Wait_Time'].mean()
daily_avg_wait_time_by_time_of_day.plot(kind = 'line')
plt.xlabel('Time of Day')
plt.ylabel('Average Wait Time')
plt.title('Average Wait Time by Time of Day')
plt.show()
```

```python
passengers_per_day_of_week = BASA['Day_of_Week'].value_counts().sort_index()
passengers_per_day_of_week
```

```python
daily_avg_wait_time_by_day_of_week
```

```python
passengers_per_day_of_week.corr(daily_avg_wait_time_by_day_of_week) # busyness 
```

```python
passengers_by_month = BASA['Month'].value_counts().sort_index()
passengers_by_month
```

```python
daily_avg_wait_time_by_month
```

```python
passengers_by_month.corr(daily_avg_wait_time_by_month) # busyness 
```
